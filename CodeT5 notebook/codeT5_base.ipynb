```python
from transformers import RobertaTokenizer, T5ForConditionalGeneration

model_name = "Salesforce/codet5-base"
tokenizer = RobertaTokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)
```

```python
from datasets import load_dataset
dataset = load_dataset("jitx/Methods2Test_java_unit_test_code")
```

```python
import pandas as pd
df_train = pd.DataFrame(dataset["train"]).head(50000)
df_val = pd.DataFrame(dataset["train"]).tail(5000)
df_test = pd.DataFrame(dataset["test"]).head(20000)
```

```python
file_path1 = "/content/sample_data/train.csv"
df_train.to_csv(file_path1, index=False)
file_path3 = "/content/sample_data/validation.csv"
df_val.to_csv(file_path3, index=False)
file_path2 = "/content/sample_data/test.csv"
df_test.to_csv(file_path2, index=False)
```

```python
from datasets import load_dataset

dsVal = load_dataset("csv", data_files={"validation": "/content/sample_data/validation.csv"})
dsTrain = load_dataset("csv", data_files={"train": "/content/sample_data/train.csv"})
dsTest = load_dataset("csv", data_files={"test": "/content/sample_data/test.csv"})
```

```python
def preprocess_function(examples):
    # Use the most detailed context as input (src_fm_fc_ms_ff)
    inputs = ["generate unit test: " + ex for ex in examples["src_fm_fc_ms_ff"]]
    targets = examples["target"]

    # Tokenize inputs and targets
    model_inputs = tokenizer(inputs, max_length=512, truncation=True, padding="max_length")
    labels = tokenizer(targets, max_length=512, truncation=True, padding="max_length")

    # Add tokenized labels to inputs
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

train_dataset = dsTrain["train"].map(preprocess_function, batched=True)
validation_dataset = dsVal["validation"].map(preprocess_function, batched=True)
train_dataset = train_dataset.remove_columns(["src_fm", "src_fm_fc", "src_fm_fc_co", "src_fm_fc_ms", "src_fm_fc_ms_ff", "target"])
validation_dataset = validation_dataset.remove_columns(["src_fm", "src_fm_fc", "src_fm_fc_co", "src_fm_fc_ms", "src_fm_fc_ms_ff", "target"])
```

```python
from accelerate import Accelerator
Accelerator().state._reset_state()
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=2,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    evaluation_strategy="steps",
    save_strategy="steps",
    logging_dir="./logs",
    learning_rate=5e-5,
    save_total_limit=2,
    logging_steps=10,
    eval_steps=1000,
    warmup_steps=500,
    fp16=True

)
```

```python
from transformers import Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=validation_dataset,
    tokenizer=tokenizer
)
```

```python
trainer.train()
```
